```python
def greeting(name):
    return 'Hi ' + name + '. How are you?'
```

```python
default_name = 'Dave'
```

```python
# nuclio:ignore
from nuclio import Context, Event
context = Context()
event = Event(body=default_name)
```

```python
# nuclio:handler
greeting(event.body)
```
